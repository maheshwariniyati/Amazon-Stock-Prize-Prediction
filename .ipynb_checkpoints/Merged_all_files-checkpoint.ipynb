```python
#importing all the necessary libraries
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
stock_data=pd.read_excel("AMZN.xlsx")
SP500=pd.read_excel("SP500.xls")
fang_data=pd.read_excel("FANG.xlsx",sheet_name='FANG')
f_f_data=pd.read_excel("F-F_Research_Data_Factors_daily.xlsx")
ads_data=pd.read_excel("ADS.xlsx")
```

```python
sp_amzn=stock_data.merge(SP500)
sp_amzn.head()
```

```python
ads_sp_amzn=sp_amzn.merge(ads_data)
```

```python
fang_data['Date'] = pd.to_datetime(fang_data.Date)
```

```python
fang_data.info()
```

```python
fang_ads_sp_amzn= ads_sp_amzn.merge(fang_data[['Date','AMZN']], how='left')
fang_ads_sp_amzn=fang_ads_sp_amzn.dropna()
fang_ads_sp_amzn=fang_ads_sp_amzn.reset_index()
fang_ads_sp_amzn=fang_ads_sp_amzn.drop('index',axis=1)
```

```python
fang_ads_sp_amzn.head()
```

```python
f_f_data['Date'] = f_f_data['Date'].astype(str)
```

```python
f_f_data['Date']
```

```python
new_data=[]
def insert_dash(data):
    for string in data:
        string = string[0:4]+'-'+string[4:6]+'-'+string[6:]
        new_data.append(string)
    return new_data
```

```python
ff_list = f_f_data['Date'].tolist()
```

```python
f_f_data['Date'] = insert_dash(ff_list)
```

```python
f_f_data.info()
```

```python
f_f_data['Date'] = pd.to_datetime(f_f_data.Date)
```

```python
df=fang_ads_sp_amzn.merge(f_f_data)
```

```python
df.head()
```

```python
df=df.rename(columns={'AMZN' : 'Returns'})
```

```python
df['Open_Close']=(df['Open ']+df['Close '])/2
df['High_Low']=(df['High']+df['Low'])/2
df['HLOC']=(df['High']+df['Low']+df['Open ']+df['Close '])/4
df['EWM_OpenClose']=df.Open_Close.ewm(span=5, adjust=False).mean()
```

```python
df = df[['Date', 'Open ','Close ', 'High', 'Low', 'Volume', 'ADS_Index',
       'Mkt-RF', 'SMB', 'HML', 'RF','Returns','SP500','Open_Close','High_Low','HLOC','EWM_OpenClose']]
```

```python
df.head()
```

```python
df.to_csv("Merged_data.csv", index=False)
```
